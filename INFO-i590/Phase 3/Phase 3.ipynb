### Import dependencies

```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
import sklearn.metrics as sm
from sklearn import datasets
from sklearn.metrics import confusion_matrix, classification_report

%matplotlib inline
```

### Import data and impute missing values

```python
data = pd.read_csv("breast-cancer-wisconsin.data", header = None, na_values = '?')
data.columns = ['scn','a2','a3','a4','a5','a6','a7','a8','a9','a10','class']
data["a7"] = data["a7"].fillna(data["a7"].median())
```

### Pare down dataset to feature set and target

```python
X = data[["a2","a3","a4","a5","a6","a7","a8","a9","a10"]]
y = data["class"]
```

# Cluster (5 pts)

### Fit model with new parameters

```python
cluster = KMeans(n_clusters = 2, n_init = 20, max_iter = 500, random_state = 42)
cluster.fit(X)
```

### Save labels

```python
labels = cluster.predict(X)
labels = np.where(labels == 0,2,4)
```

# DataFrame (5 pts)

### Add labels to dataframe

```python
X["labels"] = labels
```

### Print first 15 values

```python
X.head(15)
```

# ErrorRate Function (15 pts)

### Create ErrorRate function

```python
def ErrorRate(predicted, actual):
    frame = pd.DataFrame({'predicted': predicted, 'actual':actual})
    frame2 = frame[frame["predicted"] == 2]
    frame4 = frame[frame["predicted"] == 4]
    
    # Calculate numerator for errorB
    result1 = np.where((frame4["predicted"] != frame4["actual"]),1,0)
    frame4["result"] = result1
    countB = frame4["result"].sum()
    
    # Calculate numerator for errorM
    result2 = np.where((frame2["predicted"] != frame2["actual"]),1,0)
    frame2["result"] = result2
    countM = frame2["result"].sum()
    
    # Calculate each error rate
    errorB = countB / frame4.shape[0]
    errorM = countM / frame2.shape[0]
    total =  1-((predicted == actual).value_counts().tolist()[0] / predicted.shape[0])
    
    print("Benign error rate: " + str(round(errorB,4)))
    print("Malignant error rate: " + str(round(errorM,4)))
    print("Total error rate: " + str(round(total,4)))
```

### Test ErrorRate function

```python
ErrorRate(X["labels"], y)
```
