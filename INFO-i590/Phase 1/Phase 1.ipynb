### Import dependencies

```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
```

### Import data and add headers

```python
data = pd.read_csv("breast-cancer-wisconsin.data", header = None, na_values = '?')
data.columns = ['scn','a2','a3','a4','a5','a6','a7','a8','a9','a10','class']
```

### Impute missing data

```python
data["a7"] = data["a7"].fillna(data["a7"].median())
```

### Plot histograms for each attribute

```python
attributes = data[["a2","a3","a4","a5","a6","a7","a8","a9","a10"]]
attributes.plot(kind = "hist", edgecolor = "black", bins = 10, subplots = True, layout = (3,3), figsize = (12,20))
```

### Descriptive statistics

```python
for i in list(attributes.columns):
    print("Attribute: " + str(i))
    print("\tmean: " + str(round(attributes[i].mean(),4)))
    print("\tmedian: "+ str(round(attributes[i].median(),4)))
    print("\tStandard Deviation: " + str(round(attributes[i].std(),4)))
    print("\tVariance: " + str(round(attributes[i].var(),4)))
```
